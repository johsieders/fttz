- Write a function `faculty(n: int) -> int` which returns n!

```python
#collapse
def faculty(n: int) -> int:
    """
    :param n: integer
    :return:  nth-faculty
    standard solution
    """
    if n < 0:
        raise ValueError
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

faculty(20)
```

- Write a recursive function `faculty(n: int) -> int` which returns n!

```python
#collapse
def faculty1(n: int) -> int:
    """
    :param n: integer
    :return:  nth-faculty
    recursive solution
    """
    if n < 0:
        raise ValueError
    elif n <= 1:
        return 1
    else:
        return n * faculty1(n - 1)
    
faculty1(20)
```

- Write a function `fibo(n: int) -> int` which returns the n-th fibonacci number.

```python
#collapse
def fibo(n: int) -> int:
    """
    :param n: integer >= 0
    :return: n-th Fibonacci number
    standard solution
    """
    if n < 0:
        raise ValueError
    elif n == 0:
        return 0
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b

fibo(20)
```

- Write a recursive function `fibo1(n: int) -> int` which returns the n-th fibonacci number.
- What about stack depth and performance?

```python
#collapse
def fibo1(n: int) -> int:
    """
    :param n: integer >= 0
    :return: n-th Fibonacci number
    """
    # recursive programming, cool but slow
    if n < 0:
        raise ValueError
    elif n <= 1:
        return n
    else:
        return fibo1(n - 2) + fibo1(n - 1)

fibo1(20)
```

- Write a function `gcd(a, b: int) -> int` which returns the gcd of a and b.

```python
#collapse
def gcd(a, b: int) -> int:
    """
    :param a: integer
    :param b: integer
    :return: greatest common divisor of a and b
    standard solution
    """
    while b != 0:
        a, b = b, a % b
    return a

gcd(235, 25)
```

- Write a recursive function `gcd1(a, b: int) -> int` which returns the gcd of a and b.
- What about stack depth and performance?

```python
#collapse
def gcd1(a, b: int) -> int:
    """
    :param a: integer
    :param b: integer
    :return: greatest common divisor of a and b
    recursive solution
    """
    return a if b == 0 \
        else gcd1(b, a % b)

gcd1(235, 25)
```

- Write a function `reverse(xs: list) -> None` which reverses the order of xs on place.
Note: This is the same as `xs.reverse()`

```python
#collapse
def reverse(xs: list) -> None:
    """
    :param xs: a list
    :return: None
    this function reverses the order of xs
    standard solution
    """
    m = len(xs) // 2
    for i in range(m):
        xs[i], xs[-1 - i] = xs[-1 - i], xs[i]
        
xs = [1, 2, 3]
reverse(xs)
xs
```

- Write a function `normstring(s: str) -> str` which skips non-ascii characters and converts
the resulting string to lowercase

```python
#collapse
def normstring(s: str) -> str:
    """
    :param s: a string
    :return: keeps ascii letters only and converts s to lower case
    """
    import string
    result = ''
    for c in s:
        if c in string.ascii_letters:
            result += str.lower(c)
    return result
```

- Write a function `palindrome(s: str) -> bool` which returns True if s is a palindrome
such as this one: 'Madam, I'm Adam'.

```python
#collapse
def palindrome1(xs: str) -> bool:
    """
    :param xs: string
    :return: true if xs is a palindrome
    using Python's reverse which only works on lists
    """
    ys = list(normstring(xs))
    zs = list(ys)
    zs.reverse()
    return ys == zs

palindrome1("Madam, I'm Adam")
```

- Write a recursive function `palindrome(s: str) -> bool` which returns true if s is a palindrome
such as this one: 'Madam, I'm Adam'.

- Write a functions `romans() -> list[str]` which returns the list of the roman numbers from 1 to 4999

```python
#collapse
def romans() -> List[str]:
    """
    :return: romans numbers from 1 to 4999
    """
    digits1 = ['', 'I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX']
    digits2 = ['', 'X', 'XX', 'XXX', 'XL', 'L', 'LX', 'LXX', 'LXXX', 'XC']
    digits3 = ['', 'C', 'CC', 'CCC', 'CD', 'D', 'DC', 'DCC', 'DCCC', 'CM']
    digits4 = ['', 'M', 'MM', 'MMM', 'MMMM']

    result = []
    for d4 in digits4:
        for d3 in digits3:
            for d2 in digits2:
                for d1 in digits1:
                    result.append(d4 + d3 + d2 + d1)
    return result[1:]


romans()[:20]

```
